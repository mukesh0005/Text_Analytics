## Harnessing external resources and APIs

```python
import openai
import requests
import creds
```

```python
API_KEY = creds.news_api_key
url = f'https://newsapi.org/v2/everything?q=tesla&from=2023-09-05&sortBy=publishedAt&apiKey={API_KEY}'
```

```python
response = requests.get(url)
data = response.json()
print(response.status_code)
```

```python
print(data['articles'])
```

```python
articles = data['articles']
articles
```

```python
url_list = []
Description_list = []
title_list =[]
for article in articles:
    url_list.append(article['url'])
    Description_list.append(article['content'])
    title_list.append(article['title'])
```

```python
url_list
```

```python
import re
pattern = r"Model 3"
M3_list = []
for x in Description_list:
    match = re.search(pattern,x)
    if match:
        M3_list.append(x)
        index = Description_list.index(x)
        url = url_list[index]
        title = title_list[index]
M3_list,index,url
```

```python
# Initialize the GPT-3.5-turbo API client
openai.api_key = creds.opai_api_key
```

```python
def ask_gpt3_question(prompt):
    model_engine = "gpt-3.5-turbo-instruct"  # Replace with the engine suited for your application
    response = openai.Completion.create(
      engine=model_engine,
      prompt=prompt,
      max_tokens=200  # Limit the response to 100 tokens
    )
    
    answer = response.choices[0].text.strip()
    return answer
```

```python
# Define a sample question
sample_question = f"The {url} is an tesla article with {title}. Can you provide reviews on Model 3 "

# Use the function to ask the question
answer = ask_gpt3_question(sample_question)

# Print the answer
print(f"Answer: {answer}")
```
